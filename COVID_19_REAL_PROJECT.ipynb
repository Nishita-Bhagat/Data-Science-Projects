**COVID-19 REAL PROJECT**

We have taken a small dataset of Covid-19 but the original dataset contains more rows and columns.
The data here used is till 29-April-2020.
This data is available as a CSV file downloaded from kaggle.
We will analyze the data using Pandas dataframe.

```python
import pandas as pd
import io
  
from google.colab import files
  
  
uploaded = files.upload()
```

```python
df = pd.read_csv(io.BytesIO(uploaded['4. covid_19_data.csv']))
print(df)
```

```python
df # to see the data
```

```python
df.count() # Known values.
```

```python
df.isnull() # It will give result in Boolean form. False : No null value. True : Null value present.
# Null value : Missing number or data
```

```python
df.isnull().sum() 
```

```python
# To visualize the data
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
# To see the null values and non null values.
sns.heatmap(df.isnull())
plt.show     # Wide space or gap are the null values.
```

REMOVE ALL THE RECORDS WHERE CONFIRMED CASES ARE LESS THEN 10.

```python
df.Confirmed < 10  # To check confirmed cases less than 10.
```

```python
df[df.Confirmed < 10]  # Showing the result in form of dataset. 
```

```python
df[~(df.Confirmed < 10)] # To remove cases less then 10 from Confirmed cases. 
```

```python
df
```

```python
df = df[~(df.Confirmed < 10)] # To permanently make the changes.
```
